# 1. Deskripsi Data
Dataset ini berisi total nilai belanja Pemerintah Indonesia berdasarkan realisasi APBN pada tahun Anggaran 1990-2013 untuk pos belanja Pemerintah Pusat dan Belanja untuk daerah. Data yang tersajikan ini sudah ter-summary hinggal ke level jenis/tipe pengeluaran. Walaupun di web data.go.id menyatakan bahwa data yang tersaji dari tahun 1990-2013, secara aktual data-nya untuk tahun 2012 dan 2013 nilainya tidak terisi.

Penjelasan mengenai Variabel pada Dataset ini:

* tahun: Tahun APBN
* pos_belanja: Pos Belanja pada APBN (Belanja Pemerintah Pusat atau Anggaran Belanja untuk Daerah)
* tipe_pengeluaran: Tipe dari pengeluaran
* keterangan_pengeluaran: Keterangan mengenai pengeluaran tersebut
* nilai: Nilai dari Belanja tersebut (dalam Milyar Rupiah)

```python
import pandas as pd

real_apbn = pd.read_csv('processed-realisasi-apbn-belanja-indonesia-1990-2013.csv', delimiter = ',', encoding = "ISO-8859-1")
# real_apbn.columns
# real_apbn.iloc[10:15]
```

# 2. Statistik Data

Pada bagian ini akan disajikan beberapa statistik yang terkait dengan data, diantaranya adalah:

* Kolom yang tersedia
* Jumlah Record secara keseluruhan dan per tahun
* Nilai maksimum dan minimum dari realisasi belanja per tahun

```python
# Rata-rata realisasi belanja per tahun

real_apbn.groupby('tahun').nilai.mean()
```

```python
# TOtal realisasi belanja per tahun, pos pengeluaran dan tipe pengeluaran

real_apbn.groupby(['tahun','pos_belanja','tipe_pengeluaran']).nilai.sum()
```

# 3. Visualisasi Data

Pada bagian ini akan disajikan beberapa visualisasi yang terkait dengan data, diantaranya adalah:

* Trend Total Realisasi Belanja per Tahun periode 1990-2012
* Perbandingan Total selama periode 1990-2011 berdasar pos pengeluaran
* Perbandingan Total selama periode 1990-2011 berdasar tipe pengeluaran
* Top 3 tahun dengan jumlah realisasi belanja terbesar periode 1990-2012
* Proporsi Realisasi Pos Belanja pada tahun 2009
* Trend Realisasi Belanja per Tahun per Pos Belanja periode 1990-2012

```python
# Trend Total Realisasi Belanja per Tahun < 2012

ax = real_apbn[real_apbn['tahun']<2012].groupby('tahun').nilai.sum()
ax = ax.plot(kind='line', title ="Total Berdasarkan Tahun", marker='o', figsize = (10,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=10)
ax.set_ylabel("Realisasi", fontsize=10)
```

```python
# Perbandingan Total selama periode 1990-2011 berdasar pos belanja

ax = real_apbn[real_apbn['tahun']<2012].groupby('pos_belanja').nilai.sum()
ax = ax.plot(kind='bar', title ="Total Berdasarkan Pos Belanja periode 1990-2011", figsize = (10,5), legend=True, fontsize=12)
ax.set_xlabel("Pos Belanja", fontsize=10)
ax.set_ylabel("Realisasi", fontsize=10)
```

```python
# Perbandingan Total selama periode 1990-2011 berdasar tipe pengeluaran

ax = real_apbn[real_apbn['tahun']<2012].groupby('tipe_pengeluaran').nilai.sum()
ax = ax.plot(kind='bar', title ="Total Berdasarkan Tipe Pengeluaran periode 1990-2011", figsize = (10,5), legend=True, fontsize=12)
ax.set_xlabel("Tipe Pengeluaran", fontsize=10)
ax.set_ylabel("Realisasi", fontsize=10)
```

```python
# Top 3 tahun dengan jumlah realisasi belanja terbesar

ax = real_apbn[real_apbn['tahun']<2012].groupby('tahun').nilai.sum().nlargest(3)
ax = ax.plot(kind='bar', title ="Top 3 realisasi belanja terbesar periode 1990-2011", figsize = (10,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=10)
ax.set_ylabel("Realisasi", fontsize=10)
```

```python
# Proporsi Realisasi Pos Belanja pada tahun 2009

ax = real_apbn[real_apbn['tahun'] == 2009].groupby('pos_belanja').nilai.sum()
ax = ax.plot(kind='pie', title ="Proporsi Pos Belanja pada Tahun 2009", figsize = (7,7), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=10)
ax.set_ylabel("Realisasi", fontsize=10)
```

```python
# Trend Total Realisasi Belanja per Tahun < 2012

# ax = real_apbn.loc[(real_apbn['tahun']<2012) & (real_apbn['pos_belanja']=='Belanja Pemerintah Pusat')].groupby('tahun').nilai.sum()
# ax.plot()

ax = real_apbn.loc[(real_apbn['tahun']<2012)].pivot_table('nilai', 'tahun', 'pos_belanja', aggfunc='sum')
ax = ax.plot(kind='line', marker='o', title ="Realisasi per Pos Belanja Periode 1990-2011", figsize = (10,5), legend=True, fontsize=12)
# ax.set_xlabel("Tahun", fontsize=10)
# ax.set_ylabel("Realisasi", fontsize=10)
```

1. Pada grafik pertama terlihat trend realisasi belanja pemerintah yang selalu meningkat dari tahun 1990-2011 kecuali pada tahun 1999 mengalami penurunan yang mungkin disebabkan karena masih terkena imbas krisis ekonomi pada tahun 1998. Tren ini juga mengindikasikan bahwa kemampuan belanja pemerintah yang selalu membaik pada periode tersebut.
2. Grafik kedua memperlihatkan bahwa pos belanja terbesar secara total pada periode 1990-2011 adalah pada pos belanja pemerintah pusat. Hal ini masih mengindikasikan bahwa pemerintah pusat masih mempunyai kontrol yang lebih berpengaruh terhadap pergerakan ekonomi dan pembangunan nasional.
3. Grafik ketiga menunjukkan tipe pengeluaran Belanja rutin masih mendominasi pengeluaran APBN pada periode 1990-2011. Hal ini dapat mengindikasikan bahwa pengeluaran negara untuk pembangunan masih kurang karena secara jumlah masih selisih jauh dengan belanja rutin.
4. Tiga tahun terakhir yang terdapat pada data menyumbang nilai tertinggi Realisasi belanja APBN yang dapat mengindikasikan bahwa kemampuan belanja pemerintah semakin membaik dari tahun ke tahun.
5. Grafik kelima menampilkan proporsi belanja pusat dengan transfer ke daerah pada tahun 2009. Tidak jauh beda dengan hasil pada grafik kedua yang menunjukkan proporsi belanja pusat lebih dominan daripada transfer ke daerah.
6. Grafik terakhir menggambarkan tren pengeluaran per pos belanja dari tahun 1990-2011. Secara umum, belanja pusat dan transfer ke daerah selalu mengalami kenaikan dari tahun ke tahun, sedangkan untuk pos suspen mengalami fluktuasi dan lebih sering menyentuh angka negatif (minus)
