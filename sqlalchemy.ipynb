```python
import pymysql
import pandas as pd
from sqlalchemy import create_engine

host = '127.0.0.1'
port = "3306"
username = 'root'
password = ''
database = 'classicmodels'

engine = create_engine('mysql+pymysql://'+username+':'+password+'@'+host+':'+port+'/'+database)

def run(sql):
    df = pd.read_sql_query(sql, engine)
    return df
```

```python
sql = 'select * from orders'

run(sql)
```
