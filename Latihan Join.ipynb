## One to many relationship

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

## 1. Report the account representative for each customer.

```python
sql = """ select customernumber, customername, employeenumber, concat(firstname, ' ', lastname) salesrep from customers c 
            inner join employees e on c.salesrepemployeenumber = e.employeenumber """

run(sql)
```

## 2. Report total payments for Atelier graphique.

```python
sql = """ select customername, sum(amount) total_payments from customers c inner join payments p using(customernumber)
    where customername = 'Atelier graphique'
    group by customername
"""

run(sql)
```

## 3. Report the total payments by date

```python
sql = """ select paymentdate, sum(amount) total_payments from payments 
      group by paymentdate
"""

run(sql)
```

## 4. Report the products that have not been sold.

```python
sql = """ select p.* from products p left join orderdetails od using (productcode)
    where ordernumber is null
"""

run(sql)
```

## 5. List the amount paid by each customer.

```python
sql = """ select customername, sum(amount) total_payments from customers c inner join payments p using(customernumber)
    group by customername
"""

run(sql)
```

## 6. How many orders have been placed by Herkku Gifts?

```python
sql = """ select customername, count(1) num_orders from customers c inner join orders p using(customernumber)
    where customername = 'Herkku Gifts'
    group by customername
"""

run(sql)
```

## 7. Who are the employees in Boston?

```python
sql = """ select employeenumber, concat(firstname, ' ', lastname) employeename, city from employees e inner join offices o using(officecode)
    where city = 'Boston'
"""

run(sql)
```

## 8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

```python
sql = """ select customername, sum(amount) total_payments from customers c inner join payments p using(customernumber)
    group by customername
    having total_payments > 100000
    order by total_payments desc
"""

run(sql)
```

## Many to many relationship

## 1. List products sold by order date.

```python
sql = """ select orderdate, productname from products p 
    inner join orderdetails od using (productcode) 
    inner join orders o using (ordernumber)
    order by orderdate
"""

run(sql)
```

## 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
sql = """ select orderdate, productname from products p 
    inner join orderdetails od using (productcode) 
    inner join orders o using (ordernumber)
    where productname = '1940 Ford Pickup Truck'
    order by orderdate desc
"""

run(sql)
```

## 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
sql = """ select customername, ordernumber, sum(quantityordered * priceeach) total from customers c 
    inner join orders o using (customernumber)
    inner join orderdetails od using (ordernumber) 
    group by customername, ordernumber
    having total > 25000
    order by total 
"""

run(sql)
```

## 4. Are there any products that appear on all orders?

```python
sql = """ select productcode, count(1) num_order from orderdetails od
    group by productcode
    having num_order = (select count(1) from orders)
"""

run(sql)
```

## 5. List the names of products sold at less than 80% of the MSRP.

```python
sql = """ select productname, ordernumber, priceeach/msrp persen from products p inner join orderdetails od using (productcode)
        where priceeach/msrp < 0.8
"""

run(sql)
```

## 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
sql = """ select productname, ordernumber, priceeach, buyprice, (priceeach-buyprice)/buyprice*100 persen from products p inner join orderdetails od using (productcode)
        where (priceeach-buyprice)/buyprice >= 1
        order by persen desc
"""

run(sql)
```

## 7. List the products ordered on a Monday.

```python
sql = """
select 
    productCode,
    productName,
    orderNumber,
    orderDate,
    DAYNAME(orderDate)
from products p
    INNER JOIN orderdetails od
    USING (productCode)
    INNER JOIN orders o
    USING (orderNumber)
WHERE DAYNAME(orderDate) = 'Monday'
"""

run(sql)
```

## 8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
sql = """
SELECT 
    productCode,
    productName,
    SUM(quantityOrdered) as quantity,
    status
FROM orders o
    INNER JOIN orderdetails od
    USING (orderNumber)
    INNER JOIN products p
    USING (productCode)
WHERE status = 'On Hold'
GROUP BY productCode
"""

run(sql)
```
