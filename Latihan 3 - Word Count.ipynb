Namun, nahas bagi Barcelona, karena Messi harus ditarik keluar pada menit ke-26 setelah mengalami cedera. Tanpa Messi, skor 2-0 bertahan hingga babak pertama berakhir."

```python
# import operator

paragraf = "Barcelona berhasil merebut posisi puncak klasemen La Liga Spanyol seusai menang 4-2 atas Sevilla pada pertandingan pekan kesembilan di Stadion Camp Nou, Sabtu (20/10/2018) atau Minggu dini hari WIB. Barcelona membuka keunggulan pada menit ke-2 melalui gol yang dicetak oleh Philippe Coutinho. Lionel Messi menggandakan keunggulan Barcelona, 10 menit berselang.  Namun, nahas bagi Barcelona, karena Messi harus ditarik keluar pada menit ke-26 setelah mengalami cedera. Tanpa Messi, skor 2-0 bertahan hingga babak pertama berakhir."
print(paragraf)

#replace punctuation character and then lowercase
paragraf = paragraf.replace(',',' ').replace('/',' ').replace('-',' ').replace('(',' ').replace(')',' ').replace('.','').lower()
kata = paragraf.split()

setKata = set(kata)
listKata = list(setKata)

dictKata = {}

for i in listKata:
    kataCount = kata.count(i)
    dictKata[i] = kataCount

# sorted_dict = sorted(dictKata.items(), key=operator.itemgetter(1))
# print(sorted_dict)

dictKata_sorted = sorted(dictKata.items(), key=lambda kv: kv[1], reverse = True)
#print(dictKata_sorted)

print('\nWord Count masing-masing kata adalah sebagai berikut (urut dari yang terbanyak) :')

for l in dictKata_sorted:
    print('\u2022',l[0],':',l[1])
    
```
